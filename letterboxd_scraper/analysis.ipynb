Total Hours Watched

Difference Average vs Personal Rating

```python
films['rating'] = films['rating'].replace(['None'],100)
films
```

```python
films.to_csv('films.csv')
```

```python
films.insert(loc=6, column='diff_rating', value= 0)
```

```python
films['diff_rating'] = pd.to_numeric(films['rating'], errors='coerce') - films['average_rating']
films
```

Films Watched Per Year

```python
films.sort_values(by=['year'], ascending=True).groupby('year').count()['url']
```

Average Rating per Year

```python
films[films['numeric_rating'] <= 5].groupby('year')['numeric_rating'].mean()
```

Films Watched Per Language

```python
films.groupby('original_language').count()['url']
```

Top 10 Films Rated higher than average

```python
films[films['diff_rating'] < 5].sort_values(by=['diff_rating'], ascending=False).iloc[0:10]
```

Top 10 Films Rated lower than average

```python
films[films['diff_rating'] < 5].sort_values(by=['diff_rating'], ascending=True).iloc[0:10]
```

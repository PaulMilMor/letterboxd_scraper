Splitting values into arrays

```python
films_copy = films.copy()
films_copy['director'] = films_copy['director'].fillna("").apply(lambda x: x.split('/'))
```

```python
films_copy['country'] = films_copy['country'].fillna("").apply(lambda x: x.split('/'))
```

```python
films_copy['spoken_languages'] = films_copy['spoken_languages'].fillna("").apply(lambda x: x.split('/'))
```

```python
films_copy.insert(loc=10, column='total_languages', value=films_copy['spoken_languages'].apply(lambda x: len(x)))
```

```python
films_copy['studio'] = films_copy['studio'].fillna("").apply(lambda x: x.split('/'))
```

```python
films_copy['genres'] = films_copy['genres'].fillna("").apply(lambda x: x.split('/'))
```

```python
films_copy['themes'] = films_copy['themes'].fillna("").apply(lambda x: x.split('/'))
```

```python
films_copy['composer'] = films_copy['composer'].fillna("").apply(lambda x: x.split('/'))
```

```python
films_copy['cast'] = films_copy['cast'].fillna("").apply(lambda x: x.split('/'))
```
